```python
# Please write your code here
import urllib.request
import requests
from bs4 import BeautifulSoup
import pandas as pd
heading = []
heading_description =[] 
for number in range(50):
  review="https://www.flipkart.com/redmi-9-prime-space-blue-64-gb/p/itmddf6787a707d2?pid=MOBFUT9M84NDDYQN&lid=LSTMOBFUT9M84NDDYQN9NHPW2&marketplace=FLIPKART&srno=b_1_1&otracker=nmenu_sub_Electronics_0_Mi&fm=organic&iid=9b2d4fd7-5f04-400a-ac5b-b6885e23b017.MOBFUT9M84NDDYQN.SEARCH&ppt=pp&ppn=pp&ssid=uccf10odvk0000001614915231823"+ str(number) # Generating link dynamically
  acess = requests.get(review) 
  soup = BeautifulSoup(acess.text, 'html.parser')
  review_heading = soup.find_all(class_='_2-N8zT') 
  review_description = soup.find_all(class_='t-ZTKy') 
  for ele, sub_ele in zip(review_heading,review_description) : 
      heading.append(ele.text) 
      heading_description.append(sub_ele.text)
review_dataframe= pd.DataFrame(list(zip(heading,heading_description)), columns =['Heading', 'Heading Description'])
print("Length of data frame is {0}".format(len(review_dataframe)))
review_dataframe
```

```python
#td-idf feature

#Lower Casing

review_dataframe['After Lowercasing'] = review_dataframe['Heading Description'].apply(lambda x: " ".join(x.lower() for x in x.split()))
review_dataframe
```

```python
#td-idf feature

#Punctuational Removal

review_dataframe['After Punctuational Removal'] = review_dataframe['After Lowercasing'].str.replace('[^\w\s]','')
review_dataframe
```

```python
#td-idf feature

#Stopwords Removal
from nltk.corpus import stopwords
stop = stopwords.words('english')
review_dataframe['Stopwords Removal'] = review_dataframe['After Punctuational Removal'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
review_dataframe
```

```python
#td-idf feature

#Special Charachters
import re
review_dataframe['After removing special Charachters'] = review_dataframe['After Lowercasing'].apply(lambda x: ''.join(re.sub(r"[^a-zA-Z0-9]+", ' ', charctr) for charctr in x ))
review_dataframe
```

```python
#td-idf feature

#Removing Numbers
review_dataframe['After Removing Numbers'] = review_dataframe['After removing special Charachters'].apply(lambda x: ''.join([i for i in x if not i.isdigit()]))
review_dataframe
```

```python
#td-idf features

#Stemming
from nltk.stem import PorterStemmer
st = PorterStemmer()
review_dataframe['After Removing Numbers'].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
review_dataframe

```

```python
#td-idf feature

#Lemmatization
from textblob import Word
import nltk
nltk.download('wordnet')

review_dataframe['After Lemmatization'] = review_dataframe['After Removing Numbers'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
review_dataframe
```

```python
from nltk.tokenize import word_tokenize
pos = []
for sentence in review_dataframe['After Lowercasing']:
  text = word_tokenize(sentence)
  pos.append(nltk.pos_tag(text))
pos

Noun = []
Pronoun = []
Verb = []
Adjective = []
Adverb = []
CordinatingConjunction = []
SubordinatingConjuction = []
Interjection = []
Preposition= []
Predeterminer = []

for value in pos:
  n=0
  pp=0
  v=0
  ad=0
  av=0
  cc=0
  sc=0
  inte=0
  pr=0
  pre=0
  for word,i in value:
    if(i=='NN'):
        n=n+1
    elif(i =='PRP'):
      pp=pp+1
    elif(i =='VB'):
      v=v+1
    elif(i =='JJ'):
      ad=ad+1
    elif(i =='RB'):
      av=av+1
    elif(i =='CC'):
      cc=cc+1
    elif(i =='SC'):
      sc=sc+1
    elif(i =='IN'):
      inte=inte+1
    elif(i =='JJ'):
      pr=pr+1
    elif(i =='PDT'):
      pre=pre+1
  Noun.append(n)
  Pronoun.append(pp)
  Verb.append(v)
  Adjective.append(ad)
  Adverb.append(ad)
  Interjection.append(inte)
  Preposition.append(pr)
  Predeterminer.append(pre)
  CordinatingConjunction.append(cc)
  SubordinatingConjuction.append(sc)

review_dataframe['count of nouns'] = Noun
review_dataframe['count of Pronouns'] = Pronoun
review_dataframe['count of verbs'] = Verb
review_dataframe['count of adjectives'] = Adjective
review_dataframe['count of adverbs'] = Adverb
review_dataframe['count of interjections']= Interjection
review_dataframe['count of prepositions'] = Preposition
review_dataframe['count of predeterminers'] = Predeterminer
review_dataframe['count of CordinatingConjunction']=CordinatingConjunction
review_dataframe['count of Subordinating Conjuctions'] = SubordinatingConjuction
review_dataframe
```

```python
#Linguistic features

import spacy
RBN = []
p= spacy.load("en_core_web_sm")
for i in review_dataframe['After Lowercasing']:
  doc = p(i)
  try:
    RBN.append(doc[0].n_rights)
  except:
    RBN.append('No')
review_dataframe['Number of Right Branching Nodes across all constiuent types'] = RBN
review_dataframe['Length of sentence'] = review_dataframe['Heading Description'].apply(lambda x: len(x))
review_dataframe
```
